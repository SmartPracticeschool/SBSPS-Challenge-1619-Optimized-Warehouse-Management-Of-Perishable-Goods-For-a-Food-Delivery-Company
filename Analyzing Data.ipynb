### Determining the qualitative amount of workers needed in the centers

```python
def no_category(num):
    if num < 136.0:
        return 'Low'
    elif num < 324.0:
        return 'Medium'
    elif num < 609.0:
        return 'High'
    else:
        return 'Very High'
    
train_copy['num_orders_category'] = train_copy['num_orders'].apply(no_category)
train_copy.head()
```

```python
data =train_copy.groupby(['num_orders_category'])['num_orders'].sum().reset_index()
```
