```python
import pandas as pd 
```

```python
df=pd.read_csv('train.csv')   #read the dataframe
```

```python
df.head()                                                      #merging on dataframe
meal_info=pd.read_csv('meal_info.csv')
fulfilment_center=pd.read_csv('fulfilment_center_info.csv')
```

```python
df=pd.merge(df,meal_info,left_on='meal_id',right_on='meal_id')
df=pd.merge(df,fulfilment_center,left_on='center_id',right_on='center_id')
```

```python
df.head()
```

```python
df.drop(columns=['id','city_code','region_code'],inplace=True)
```

```python
df=pd.get_dummies(df,columns=['center_type','cuisine'])
```

```python
y=df['category']  #target variable
```

```python
x=df.drop('category',axis=1)
x.head()
```

```python
# training a DescisionTreeClassifier 
from sklearn.tree import DecisionTreeClassifier 
dtree_model = DecisionTreeClassifier(max_depth = 10).fit(x, y) 
dtree_predictions = dtree_model.predict(x) 
  
# creating a confusion matrix 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, dtree_predictions)
```
